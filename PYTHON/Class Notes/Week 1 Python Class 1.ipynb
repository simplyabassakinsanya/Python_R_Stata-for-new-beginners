```python
## Variables & Data Types ##
### Lists - ordered, changeable ### 
fruits = ["apple", "banana", "cherry"]
print(fruits [0]) # access first item 
fruits.append ("orange") # add new item
print(fruits)
```

```python
## Dictionaries - key value pairs ##
student = { "name": "Abass", "age": 20, "score": 85}
print(student["name"]) #access value by key
student["score"] = 90 #update value
print(student)
```

```python
## Control Flow: if/else & Loops ##
### If/Else ###
score = 75
if score >= 50: 
    print ("pass")
else:
    print ("fail")
```

```python
## Control Flow: if/else & Loops ##
### If/Else ###
score = 45
if score >=50: 
    print ("pass")
else:
    print ("fail")
```

```python
## For Loop: Repeat for each item in a group ##
for fruit in ["apple", "banana", "cherry"]:
    print (fruit)
```

```python
## While Loop: Repeat until a condition is no longer true ###
count = 1
while count <=3:
    print("hello", count)
    count += 1
```

```python
## Fucntions: Reusable Code Blocks ##
def greet (name):
    return f"Hello, {name}!"
print (greet ("Abass"))
print (greet ("Damola"))
```
