```python
# Conditional Statement

a = 300
b = 23
if a > b:
    print("a is greater than b")
```

```python
# Elif Keyword
a = 33
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")

# elif means "else if"
# If the first code is not true, elif will be activated but if its true, elif will not be activated/run
```

```python
a = 200
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
```

```python
a = 200
b = 33
if b > a:
    print("b is greater than a")
else:
    print("b is not greater than a")
```

```python
# Short hand if...else
# if you have only one statement to execute, one for if, and one for else, you can put it all on the same line.

a = 30
b = 20
if a > b: print("a is less than b")
```

```python
a = 2
b = 330

print("A") if a > b else print ("B") # Note: This technique is known as Ternary Operators or Conditional Expressions.
```

```python
# You can not use elif for the short hand. You can use multiple elif
# One line if else statement, with 3 conditions.

a = 330
b = 330

print("A") if a > b else print("=") if a == b else print("B")
```

```python
# And keyword
# The and keyword is a logical operator, and is used to combine conditional statements:

# Test if a is greater than b, AND if c is greater than a:
a = 200
b = 33
c = 500
if a > b and c > a:
    print("Both conditions are True")
```

```python
# Or operator: Only one condition needs to be true, not necessary that all conditions have to be true

a = 200
b = 33
c = 500
if a > b or a < c:
    print("At least one of the conditions is True")
```

```python
# Not operator: This is used to reverse the result of the conditional statement.
# Test is a is not greater than b:
a = 33
b = 200
if not a > b:
    print("a is not greater than b")
```

```python
# Nested if: Used if 2 conditions were to be true at the same time.

x = 41

if x > 10:
    print("Above ten")
    if x > 20:
        print("and also above 20")
    else:
        print("but not above 20")
```
