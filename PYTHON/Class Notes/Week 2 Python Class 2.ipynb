```python
# How to fill all columns at once

# If you want to fill missing values in every numeric column with its mean:

# data.mean(numeric_only=True) calculates the mean for each numeric column.

# fillna(...) replaces missing values in those columns with their mean.

# Now isna().sum() should show 0 for all numeric columns.
```

```python
# Removing Duplicates

# Remove duplicate rows
datasets = datasets.drop_duplicates()
print(datasets)
```

```python
datasets['Age '] = pd.to_numeric(datasets['Age '], errors="coerce")

print(datasets.isna().sum())
```

```python
# Changing Data Types

# Convert Age to numeric (turns "twenty" into NaN)

# python data["Age"] = pd.to_numeric(data["Age"],errors="coerce") Problem: In the dataset, Charlie's age is written as "twenty" (text),

# Why it matters: Pandas thinks "twenty" is a string, not a number. That means you can't calculate averages or do math with it.

# Solution: pd.to_numeric(...,errors="coerce") tries to turn everything into a number.

# If it succeeds (like "twenty"), it replaces it with NaN (Not a number).
```

```python
print(datasets.columns)
```

```python
datasets = datasets.fillna(datasets.mean(numeric_only=True))
print(datasets.isna().sum())
```

```python
## to_numeric ensures numbers are treated as numbers.

## astype(str) ensures names are text
# Convert a column to numeric
datasets["Math"] = pd.to_numeric(datasets["Math"], errors="coerce")

# Convert a column to string
datasets["Name"] = datasets["Name"].astype(str)
```

```python
# Adding New Columns

# Create a Total and Average score

# You can create new features (columns) from existing ones. This is called feature engineering.

#data["Total"] = data["Math"] + data["English"] + data["Science"]data["Math"] - selects the Math column from the Dataframe
#data["English"] - selects the English Column
#data["Science"] - selects the Science column
#Adding them together(+) - adds the scores for each student across the three subjects
#data["Total"]=...- creates a new column called Total and stores the result there.
# ----data["Average"] = data["Total"]/3 data["Total"] - the new column we just created.
# /3 - divides the total score by 3 (because there are 3 subjects)
# data["Average"] = ...-creates another new column called Average and stores the result data["Total"]=data["Math"]+data["English"]+data["Science"]data["Average"]=data["Total"]/3
```

```python
datasets["Total"] = datasets["Math"] + datasets["English"] + datasets["Science"]
datasets["Average"] = datasets["Total"] / 3

print(datasets[["Name" , "Math" , "English" , "Science" , "Total" , "Average"]])
```

```python
## Filtering Data to concentrate on a particular set
# Filtering lets you focus on speciifc groups (e.g., top performers).
# Students with Math >= 85
high_math = datasets[datasets["Math"] >= 85]
print(high_math)
```

```python
# Students with Average >= 80
strong_students = datasets[datasets["Average"] >= 80]
print(strong_students)
```
