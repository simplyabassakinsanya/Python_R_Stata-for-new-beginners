```python
# Handling Categorical Data
# Categorical data = text labels like "Male/Female", "Yes/No", "Class A/B/C".

# Computers can not do math on text, so we can convert categories into numbers.
# How we handle it:
# One-Hot Encoding (Dummy Variables):
# Each category becomes its own column with 1 or 0.
# Example: "Class A/B/C" - Class_A, Class_B, Class_C.
```

```python
# Machine learning and statistics need numbers, not text.
# 1. Categorical Variables
# Meaning: Variables that represent categories or labels (e.g., Gender = Male/Female, Department = Sales/Support).
# Use: Helps group and compare data.
# Challenge: Computers can't calculate with text
# Solution: Convert to numbers using dummy variables (One-Hot Encoding)

# 2. Dummy Variables
# Meaning: Extra columns created for each category, marked with 1 if present, 0 if not
# Use: Makes categorical data usable in analysis and visualization.
# Example:

#code
ProductCategory_Electronics ProductCategory_Fashion ProductCategory_Groceries
1                           0                       0
0                           1                       0
# Challenge: Too many categories - too many columns
```
