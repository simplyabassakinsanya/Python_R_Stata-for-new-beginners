```python
# Normal python list
x = [1,2,3,4]
print(x * 2)

# NumPy array
import numpy as np
y = np.array([1, 2, 3, 4])
print(y * 2)
```

```python
import numpy as np
# create an array
scores = np.array([85, 90, 78, 92])
# Do math on the array
print(scores.mean()) # average score
print(scores.max()) # highest score
print(scores + 5) #add 5 to every score
```

```python
x = [3,8,12,10]
import numpy as np
y = np.array(x)
print ("mean value:",y.mean())
print ("max value:",y.max())
print ("min value:",y.min())
print ("power 2 value:",y**2)
print ("minus",y-3)
print ("add",y+3)
print ("multiplied values:",y*3)
```
