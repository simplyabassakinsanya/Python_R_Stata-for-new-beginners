```python
Matplotlib and Seaborn Concept: These libraries are for data visualization

Why it matters: ML is not just about numbers - graphs helps us understand trends and patterns

Analogy: Think of Matplotlib as the "paintbrush" and Seaborn as the "artist" that makes prettier charts.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Simple plot 
scores = [85, 90, 78, 92]
plt.plot(scores)
plt.title("student scores")
plt.show()

# Seaborn for nicer visuals 
sns.histplot(datasets["Math"], bins=5, kde=True)
plt.title("Distribution of Math Scores")
plt.show
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line plot of scores
plt.plot(datasets["Name"], datasets["Math"], label="Math")
plt.plot(datasets["Name"], datasets["English"], label="English")
plt.plot(datasets["Name"], datasets["Science"], label="Science")
plt.legend()
plt.title("Student Scores by Subject")
plt.show()

# Histogram of Math scores
sns.histplot(datasets["Math"], bins=5, kde=True)
plt.title("Distribution of Math Scores")
plt.show()
```
