```python
def greet_user(name):
    print(f"Hello, {name}! Welcome to Python...") #the 'f' is the formatted string, which helps you to format the string instead of seeing it as a normal text.

greet_user("Segun")
```

```python
def my_function(first_name):
    print("Hi " + first_name)

my_function("Renah")
my_function("Favour")
my_function("Miracle")
```

```python
def greet_user(name): #'name' is a the parameter
    print(f"Thank you {name} for investing with us!")
#Anytime we are using the formatted text and you want to call the parameter, you will the curly bracket

greet_user("Comfort")
greet_user("Abass")
greet_user("Damola")
```

```python
#You can have multiple parameters, separated by commas

def add_numbers(a, b):
    return a + b

result = add_numbers(20, -13)
print(result)
```

```python
#You can have multiple parameters, separated by commas

def multiply_numbers(a, b):
    return a * b

result = multiply_numbers(20, -13)
print(result)
```

```python
# In the example below, this function expects 2 arguments, and gets 2 arguments:

def my_function(fname, lname):
    print(fname + " " + lname)

my_function("Comfort", "Ajanlekoko")
```

```python
#Bar Charts Without Using Functions

import matplotlib.pyplot as plt

#My data
students = ["Ade", "Bola", "Chidi", "Zainab"]
scores = [75, 85, 69, 90]

# My plot
plt.bar(students, scores)
plt.title("student Scores")
plt.xlabel("students")
plt.ylabel("Scores")
plt.show()
```

```python
def plot_bar_chart(students, scores, title, xlabel, ylabel):
    plt.bar(students, scores)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
plot_bar_chart(students, scores, "Student Scores", "Students", "Scores")
```
