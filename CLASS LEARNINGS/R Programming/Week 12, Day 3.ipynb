```r
# R Data Structures: This is basically how data are stored in R. Data structures are used to store and organize values. R provides many built-in data structures. Each is used to handle data in different ways.
# Examples of Data structures includes; vectors, list, matrices, arrays, and data frames.
```

```r
# R Vectors: A vector is simply a list of items that are of the same types.
# exmaple of vector
fruits <- c("banana", "apple", "orange") # Note: In R, c() means "combine" or "concantenate"
                                         # It is used to put multiple values together into a single vector.
                                         # All the values inside c() must be the same type. (Else, R will convert them)
fruits
```

```r
# Example of R converting values if not put in c()
b <- c(1, "A", TRUE)
b
# Note: Everything becomes a character.
```

```r
# Vector of numerical values
numbers <- c(1, 2, 3)
numbers
```

```r
# To create a vector with numerical values in a sequence, use the : operator
numbers <- 1:10
numbers
```

```r
# You can also create numerical values with decimals in a sequence, but note that if the last element does not belong to the sequence, it is not used:
numbers1 <- 1.5:6.5
numbers1

numbers2 <- 1.5:6.3
numbers2
```

```r
# We can create a vector of logical values:
log_values <- c(TRUE, FALSE, TRUE, FALSE)
log_values
```

```r
# Vector length can also be used to find out how many items a vector has, use the length() function:
fruits <- c("banana", "apple", "orange")
length(fruits)
```

```r
# Sort a vector to sort items in a vector alphabetically or numerically, use the sort() function:
fruits <- c("banana", "apple", "orange", "manago", "lemon")
numbers <- c(13, 3, 5, 7, 20, 2)

sort(fruits) # Sort a string
sort (numbers) # Sort numbers
```

```r
## Accessing Vectors: You can access the vector items by referring to its index number inside brackets[]. The first item has index 1, second item has index 2, and so on.
fruits <- c("banana", "apple", "orange")
fruits[1]
```

```r
# You can also access multiple elements by referring to different index positions with the c():
fruits <- c("banana", "apple", "orange", "manago", "lemon")
# Access the first and third item (banana and orange)
fruits[c(1, 3)]
```

```r
# You can also use negative index numbers to access all items except the ones specified:
fruits <- c("banana", "apple", "orange", "mango", "lemon")
fruits[c(-1)]
fruits[-4]
fruits[-1]
```

```r
# Change an item to change the value of a specific item, refer to the index number:
fruits <- c("banana", "apple", "orange", "mango", "lemon")
# change "banana" to "pear"
fruits[1] <- "pear"
fruits
fruits[c(2, 4)] <- c("pear", "guava")
fruits
```

```r
# Repeat Vectors: To repeat vectors, we use the rep() function
# rep() means repeat values in a vector.
repeat_each <- rep(c(1,2,3), each = 3) # Note: Repeat each individual element three times before moving to the next
repeat_each
```

```r
# Repeat the sequence of the vendor. Example
repeat_times <- rep(c(1,2,3), times = 3)
repeat_times
```

```r
# Repeat each value independently.
repeat_indepent <- rep(c(1,2,3), times = c(5,2,1)) # This type of repetition is called element-wise repetition.
                                                   # This is because each value is repeated independently based on its corresponding count.
                                                   # Each number repeats based on its own instruction.
repeat_indepent
```

```r
# Generating Sequenced Vectors
# To make bigger or smaller steps ina  sequence, use the seq() function:
numbers <- seq(from = 0, to = 100, by = 20) # Note: by controls the step size
numbers
# Note: The seq() function has three parameters: from is where the sequence startes, to is where the sequence stops, and by is the interval of the sequence.
```
