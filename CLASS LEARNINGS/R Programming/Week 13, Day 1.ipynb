```r
# R List
# A list in R can contain many different types inside it. A list is a collection of data which is ordered and chnageable.

thislist <- list("apple", "banana", "cherry")
thislist
```

```r
# You can access the list items by referring to its index number, inside brackets. the 1st item has index 1, 2nd item has index 2, and so on.
thislist <- list("apple", "banana", "cherry")
thislist[2]
```

```r
# Changing item value: To change the value of a specific item, refer to the index number:
thislist <- list("apple", "banana", "cherry") # remember the c() spoken about yesterday, it does not necessarily mean vector
                                              # it means to combine or concantenate. So, it will still be brought in even though this is a list.
thislist[1] <- "blackcurrant"
thislist
```

```r
# Changing multiple item value: To change the value of a multiple items, refer to the index numbers:
thislist <- list("apple", "banana", "cherry")
thislist[c(1,2)] <- list("blackcurrant", "guava")
thislist
```

```r
# List length to find out how many items a list has, use the length() function
thislist <- list("apple", "banana", "cherry")
length(thislist)
```

```r
# To check if an item exist: To find out if a specified item is present in a list, use the %in% operator.
thislist <- list("apple", "banana", "cherry")
"apple" %in% thislist
```

```r
# Add llist items: To add an item to the end of the list, use the append() function.
thislist <- list("apple", "banana", "cherry")
append(thislist, "orange")
```

```r
# To add an item to the right of a specified index, add "after=index number" in the append() function.
thislist <- list("apple", "banana", "cherry")
append(thislist, "guava", after = 2)
append(thislist, "guava", after = 1)
append(thislist, "guava", after = 0)
```

```r
# Remove list items: You can also remove list items. The following example creates a new, updated list without an "apple" item
thislist <- list("apple", "banana", "cherry")
newlist <- thislist[-1]
newlist
```

```r
# Range of indexes: You can specify a range of indexes by specifying where to start and where to end the range, by using the : operator
thislist <- list("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
(thislist)[2:5]
```

```r
# Loop thorugh a list: you can loop through the list items by using a for loop:
thislist <- list("apple", "banana", "cherry")
for (x in thislist) {
  print(x)
}
```

```r
# Join two lists: There are several ways to join or concatenate, two or more lists in R
# the most common way is to use the c() function, which combines the teo elements together:
list1 <- list("a", "b", "c")
list2 <- list(1,2,3)
list3 <- c(list1,list2)
list3
```
