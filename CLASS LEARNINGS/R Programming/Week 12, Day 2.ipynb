# Write a function that takes a score and returns pass if the score is greater than or equal to 50 and fail if the score is less than 50

```r
class_work <- function(score) {
 if (score >= 50) {
    return("pass")
} else {
    return("fail")
}
}
```

```r
class_work(50)
class_work(30)
```

# Write a function that checks if a number is positive or, negative or 0

```r
stressing_me <- function(num) {
  if (num > 0) {
    print("positive")
  } else if (num < 0) {
    print("negative")
  } else {
    print("Zero")
  }
}
```

```r
stressing_me(20)
stressing_me(-40)
stressing_me(0)
```

# Investment Return Decision: If the ROI is 15% and above, the result should be good investment, and if it is between 5% and 14%, it is average investment, and if it is below it, it is a bad investment.

```r
Investment_decision <- function(investment) {
  if (investment >= 15) {
    print("go ahead")
  } else if (investment < 5) {
    print("do not do it")
  } else {
    print("average investment")
  }
}
```

```r
Investment_decision(30)
Investment_decision(11)
Investment_decision(2)
```

```r
investment_decision <- function(roi) {
  if (roi >= 15) {
    return(paste0(roi, "% - Good Investment")) # The zero "0" means zero space between the values
  } else if (roi >= 5) {
    return(paste0(roi, "% - Average Investment"))
  } else {
    return(paste0(roi, "% - Bad Investment"))
  }
}

investment_decision(4)

# Note: paste0() is a function in R used to join or concatenate values together into one string
# without adding any space between them
```

```r
# Savings Growth Simulation
balance <- 1000
year <- 1

while (balance < 2000) {
  balance <- balance * 1.10
  print(paste("Year", year, "Balance:", balance))
  year <- year +1
}
```
