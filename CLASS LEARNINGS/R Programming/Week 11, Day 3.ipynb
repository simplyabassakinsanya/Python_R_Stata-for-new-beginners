#If...else Combined with a For Loop.

#To demonstrate a practical example, let us say we play a game of Yahtzee.

```r
dice <- 1:6
for (x in dice) {
  if (x == 6) {
    print(paste("The dice number is", x, "Yahtzee"))
  } else {
    print(paste("The dice number is", x, "Not Yahtzee"))
  }
}
```

# Nested Loops: It is also possible to place loop inside another loop. This is called a nested loop.

```r
adj <- list("red", "big", "tasty")

fruits <- list("apple", "banana", "cherry")
  for (x in adj) {
    for (y in fruits) {
      print(paste(x, y))
    }
  }
```

# RF WHILE LOOPS
# Loops can execute a block of code as long as specified condition is reached. Loops are handy because they save time, reduce errors, and they make code more readable.
# R has to loop commands: While Loop and For Loop.

```r
# Example: print i as long as i is less than 6
i <- 1
while (i < 6) {
  print(i)
  i <- i + 1
}
```

```r
# Break: With the break statement, we can stop the loop even if the while condition is true.
# Example: Exit the loop if i is equal to 4

i <- 1
while (i < 6) {
  print(i)
  i <- i + 1
  if (i == 4) {
    break
  }
}
```

```r
i <- 0
while (i < 6) {
  i <- i + 1
  if (i == 3) {
    next
  }
  print(i)
}
```
