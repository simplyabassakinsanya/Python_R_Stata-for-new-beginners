For Loops in R Programming
* For Loops: A for loop is used in iterating over a sequence.

```r
#Example

for (x in 1:10) {
  print(x)
}
```

```r
# Example: Print every in a list
# List example
fruits <- list("apple", "banana", "cherry")
for (x in fruits) {
  print(x)
}
```

```r
#Example using vector
fruits <- c("apple", "banana", "cherry")
for (x in fruits) {
  print(x)
}
```

```r
# Array version example

fruits <- array(c("apple", "banana", "cherry")) # Array () creates an array object
                                                # c() combines the values into the array
# Output for the array
fruits

for (x in fruits) {
  print(x)
}
```

```r
dice <- c(1, 2, 3, 4, 5, 6)

for (x in dice) {
  print(x)
}
```

The Break Statement: We can stop the loop before it has looped through all the items.

```r
# Example: Stop the loop at "Cherry"

fruits <- list("apple", "banana", "cherry")
for (x in fruits) {
  if (x == "cherry") {
    break
  }
  print(x)
}
```

Next: With the next statement, we can skip an iteration without terminating the loop:

```r
# Example
# Skip "Banana"

fruits <- list("apple", "banana", "cherry")
for (x in fruits) {
  if (x == "banana") {
    next
  }
  print(x)
}
```
