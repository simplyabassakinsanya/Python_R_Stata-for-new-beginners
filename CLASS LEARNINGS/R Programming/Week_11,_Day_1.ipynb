```r
# 1. Positive, Negative, Zero

num <- -3

if (num > 0) {
  print("Positive")
} else if (num < 0) {
  print("Negative")
} else {
  print("Zero")
}
```

```r
# Asking the user to input a number
num <- as.numeric(readline(prompt = "Enter a number: "))

# check for the condition
if (num > 0) {
  print("The number is Positive")
} else if (num < 0) {
  print("The number is Negative")
} else {
  print("The number is Zero")
}
```

```r
#Example 2

num <- as.numeric(readline(prompt = "Enter a number: "))
if (num >= 70) {
  print("Grade A")
} else if (num >= 60) {
  print("Grade B")
} else if (num >= 50) {
  print("Grade C")
} else if (num >= 40) {
  print("Grade D")
} else {
  print("Grade F")
}
```

#Create a discount calculator:
a) if amount is greater than N100,000 - 10% discount
b) otherwise, - No discount


```r
Amount <- 200000
if (Amount %% 10 >= 100000) {
  print("discount")
} else {
  print("No discount")
}
```

```r
# Ask user for purchase amount
amount <- as.numeric(readline(prompt = "Enter purchase amount"))
if (amount > 100000) {
  discount <- amount * 0.10
  final_price <- amount - discount

  print(paste("Discount Applied:", discount))
  print(paste("Final price: ", final_price))
} else {
  print("No Discount Applied")
  print(paste("Total amount to pay:", amount))
}
```
