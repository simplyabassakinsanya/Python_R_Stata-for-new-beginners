# R Function: A function is a blockof code which only runs when it is called.

# You can pass data, known as parameters, into a function

# A function can return data as a result.

# In R, a function is simply a reusable block of code that performs a specific task when you call (run) it.

# Think of it like a machine: You give it input (called parameters or arguments), processes the input, and it gives you an output (called the return value).

# Functions helps to avoid repeating code. They help make the code cleaner, and easier to maintain.

# Functions can take input and return output

# How to create a function using the function() keyword.
function_name <- function(parameters) {
  # code to execute
  return(result)
}

```r
my_function <- function() { # create a function with the name my_function
  print("Hello World!")
}
```

```r
my_function() # Calling out the function created
```

# Parameters and Arguments

# Parameters are the variables listed in the function definition. They act as placeholders for the values the function will recieve.

# Arguments: Arguments are specified after the function name, inside the paranthesis. Information can be passed into functions as arguments.

# Arguments are specified after the function name, inside the paranthesis. You can also add as many arguments as you want, just seperate them with a coma.

```r
# The following example has a function with one argument (fname). When the function is called, we pass along a first name, which is used inside the function to print the full name:
# Function definition (Parameters)
add_numbers <- function(a, b) {
  result <- a + b
  return(result)
}
# a and b are the parameters
```

```r
# Function Call (Arguments)
add_numbers(5, 4)
# 5 and 4 are the arguments
```

```r
my_function <- function(fname) {
  paste(fname, "Griffin") # The paste() joins strings together
}

my_function("Peter")
my_function("Lois")
my_function("Stewie")
# "Peter", "Lois", and "Stewie" are arguments
# They are the actual values
# They get assigned to fname
```

```r
# Example
# This function expects 2 arguments, and gets 2 arguments:
my_function <- function(fname, lname) {
  paste(fname, lname)
}
my_function("Peter", "Griffin")
```

# Write a function that checks if a number is even or odd.

```r
check_even_odd <- function(num) {
  if (num %% 2 == 0) {
      return("Even")
  } else {
    return("Odd")
  }
}
```

```r
check_even_odd(10)
check_even_odd(7)
```
